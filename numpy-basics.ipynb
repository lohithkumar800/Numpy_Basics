***Create a NumPy array***

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print(arr)

print(type(arr))
```

**0-array***

```python
import numpy as np

arr = np.array(42)

print(arr)
```

**2-d array**

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr)
```

**3-d array**

```python
import numpy as np
arr = np.array([[[1,2,3], [4,5,6]], [[1,2,3], [4,5,6]]])
print(arr)
```

***number of dimensions***

```python
import numpy as np
arr = np.array([1, 2, 3, 4], ndmin=5)
print(arr)
print('number of dimesions :',arr.ndmin)
```

***arrange***

```python
arr = np.arange(12)
print(arr)
```

**Reshape**

```python
arr = arr.reshape(3, 4)
print(arr)
```
